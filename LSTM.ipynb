# Time series forcasting

```python
# import  a time series data

!pip install yfinance
import yfinance as yf
data = yf.download('AAPL', start="2021-01-01", end="2023-06-30")
```

```python
# lstm for time series forcasting with predction and visualization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
# Data Preprocessing

data = data.reset_index()
data = data[['Date', 'Close']]
data['Date'] = pd.to_datetime(data['Date'])

train_data = data[data['Date'] < '2023-04-01']
test_data = data[data['Date'] >= '2023-04-01']

scaler = MinMaxScaler()
train_data_scaled = scaler.fit_transform(train_data['Close'].values.reshape(-1, 1))
test_data_scaled = scaler.transform(test_data['Close'].values.reshape(-1, 1))
```

```python
train_data.head()
```

```python
# Describe the training data
print(train_data.describe())

# Describe the test data
print(test_data.describe())
```

```python
# Visualize the data
plt.figure(figsize=(12, 6))
plt.plot(train_data['Date'], train_data['Close'], label='Train Data')
plt.plot(test_data['Date'], test_data['Close'], label='Test Data')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Apple Stock Price')
plt.legend()
plt.show()
```

```python
# LSTM Model

model = Sequential()
model.add(LSTM(500, return_sequences=True, input_shape=(train_data_scaled.shape[1], 1)))
model.add(LSTM(10))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
```

```python
model.fit(train_data_scaled, train_data_scaled, epochs=20, batch_size=16)
```

```python
# Prediction and Visualization

predictions = model.predict(test_data_scaled)
predictions = scaler.inverse_transform(predictions)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(test_data['Date'], test_data['Close'], label='Actual')
plt.plot(test_data['Date'], predictions, label='Predicted')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

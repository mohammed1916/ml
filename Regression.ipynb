# Regression



## Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets
```

## Load Dataset

```python
diabetes = datasets.load_diabetes()

X = diabetes.data
y = diabetes.target
target_names = diabetes.feature_names
```

## Data Exploration

```python
print(X.shape,y.shape)
```

```python
import pandas as pd
data = pd.DataFrame(X)
data.head()
```

```python
data.describe()
```

```python
target_names
```

## Regression

```python
# Import the necessary libraries
from sklearn.linear_model import LinearRegression

# Create a LinearRegression object
regressor = LinearRegression()

# Fit the model to the data
regressor.fit(X, y)

# Make predictions on the data
predictions = regressor.predict(data)

# Calculate the mean squared error
mse = np.mean((predictions - y)**2)

# Print the mean squared error
print("Mean squared error:", mse)



```

```python
# Fit the model to the data
regressor.fit(X, y)

# Calculate the R-squared of the model
r_squared = regressor.score(X, y)

# Print the R-squared
print("R-squared:", r_squared)
```

```python
import matplotlib.pyplot as plt

# Sort the data by the feature variable
sorted_data = np.sort(X, axis=0)

# Make predictions on the sorted data
sorted_predictions = regressor.predict(sorted_data)

# Trim the longer array
sorted_data = sorted_data[:len(y)+1]
# Pad the shorter array
#target = np.pad(y, (0, len(sorted_data) - len(y)), 'edge')
```
